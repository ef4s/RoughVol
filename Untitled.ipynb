# Volatility Is Rough

## Fractional Brownian Motion

Fractional Brownian Motion (fBM) is like Brownian Motion, but the increments are not necessarily independent. For our purposes, the most important result is that the sign of the autocorrelation is determined by the Hurst exponent:
* $H<0.5$ increments are *positively* correlated
* $H=0.5$ increments are independent
* $H>0.5$ increments are *negatively* correlated

[Man Group Data](http://realized.oxford-man.ox.ac.uk/data/download)

```python
man = pd.read_csv('OxfordManRealizedVolatilityIndices.csv', header=2, index_col=0, parse_dates=True, usecols=[0,1])
man = np.log(man['SPX2.rv'])
```

```python
fig = plt.figure()
fig.suptitle('MAN Group S&P 500 5 minute realised variance', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)
man.plot(ax=ax)

ax.set_ylabel('Log Realised Variance')
ax.xaxis.label.set_visible(False)
```

$$\log\sigma_{t+\Delta} - \log\sigma_{t}=v(W_{t+\Delta}^H - W_{t}^H)$$

```python
n = 100
exponents = [0.1, 0.25, 0.5, 0.75]
for hurst in [0.1, 0.25, 0.5, 0.75]:
    fbmr, _, times = fbm.fbm(n, H=hurst, L=1, method='daviesharte', random_seed=1337)
    plt.plot(times, fbmr)
plt.legend(exponents)
```
