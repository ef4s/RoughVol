# Volatility Is Rough

## Fractional Brownian Motion

Fractional Brownian Motion (fBM) is like Brownian Motion, but the increments are not necessarily independent. For our purposes, the most important result is that the sign of the autocorrelation is determined by the Hurst exponent:
* $H<0.5$ increments are *positively* correlated
* $H=0.5$ increments are independent
* $H>0.5$ increments are *negatively* correlated

[Man Group Data](http://realized.oxford-man.ox.ac.uk/data/download)

````python
man = pd.read_csv('OxfordManRealizedVolatilityIndices.csv', header=2, index_col=0, parse_dates=True, usecols=[0,1])['SPX2.rv']
man.dropna(inplace=True)
# man = pd.Series(man)
man_log_var = np.log(man)
man_log_std = np.log(np.sqrt(man))
# man = np.log(man['SPX2.rv'].apply(np.sqrt))
````

````python
fig = plt.figure(figsize=(8*3, 6))
fig.suptitle('Sample Paths from a Fractional Brownian Motion', fontsize=14, fontweight='bold')



n = man_log_var.size
exponents = [0.25, 0.55, 0.75]
# exponents = np.arange(0.1, 0.9, 0.1)
for i in range(len(exponents)):
    hurst = exponents[i]
    ax = fig.add_subplot(1, len(exponents), i + 1)
#     ax = fig.add_subplot(131)
    ax.set_title('Hurst Parameter: ' + str(hurst))
    ax.set_ylabel('Sample Value')
    ax.xaxis.label.set_visible(False)
    fbmr, _, times = fbm.fbm(n, H=hurst, L=1, method='daviesharte', random_seed=1337)
    plt.plot(times, fbmr, alpha=0.8)
# plt.legend(exponents)
plt.savefig('sample_paths')
````

$m(q, \Delta)=\frac{1}{N}\sum_{k=1}^N |{\log(\sigma_{k\Delta}) - \log(\sigma_{(k-1)\Delta})}|^q$

Which in python becomes:
```python 
m = sigma.log().diff(periods=k).abs().pow(q).mean()
```

````python
q_s = np.arange(0.5, 3.5, 0.5)
idx = list(range(1, int(np.exp(4))))
res = []
for q in q_s:
    _res = []
    for k in idx:
        _res.append(man_log_std.diff(periods=k).abs().pow(q).mean())
    res.append(pd.Series(_res, index=idx))
res = pd.DataFrame(res, index=q_s).T
````

### What is q?
>The basic Hurst exponent can be related to the expected size of changes, as a function of the lag between observations, as measured by $E(|X_{t+}-X_{t}|^2)$. For the generalized form of the coefficient, the exponent here is replaced by a more general term, denoted by q. [Source](https://en.wikipedia.org/wiki/Hurst_exponent)

````python
logres = res.apply(np.log)
logres.index = np.log(res.index)
````

````python
regression_coefs = {}
for l in logres:
    regr = linear_model.LinearRegression()
    x = logres[l].index.values.reshape(-1, 1)
    y = logres[l].values.reshape(-1, 1)
    regr.fit(x,y)
    regression_coefs[l] = np.array([regr.intercept_, regr.coef_[0]]).T[0]
````

````python
fig = plt.figure(figsize=(8, 6))
fig.suptitle('MAN Group S&P 500 5 minute realised variance', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)

logres.plot(style='.', colormap='viridis', ax=ax)

for l in logres:
    regr = linear_model.LinearRegression()
    x = logres[l].index.values.reshape(-1, 1)
    y = logres[l].values.reshape(-1, 1)
    regr.fit(x,y)
    plt.plot(x, regr.predict(x), color='red', linewidth=1)

# ax.xaxis.set_major_formatter(ScalarFormatter())
# ax.yaxis.set_major_formatter(ScalarFormatter())
````

$$\mathbb{E}[|{\log(\sigma_{k\Delta}) - \log(\sigma_{(k-1)\Delta})}|^q]=K_q\Delta^{\zeta_q}$$
Where $\zeta_q$ is the slope of the lines above, and it turns out that $\zeta_q \tilde{} Hq$

````python
q_regr = linear_model.LinearRegression()
q_regr.fit(np.array(q_s).reshape(-1, 1),np.array(list(regression_coefs.values()))[:, 1].reshape(-1, 1))
````

````python
fig = plt.figure(figsize=(8, 6))
fig.suptitle('$\zeta_q$ is linear in $q$', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)

plt.plot(np.array(q_s), np.array(list(regression_coefs.values()))[:, 1])
plt.plot(np.array(q_s), q_regr.predict(np.array(q_s).reshape(-1, 1)))
ax.set_ylabel('$\zeta_q$')
ax.set_xlabel('$q$')
````

This gives an estimate of $H=$ {{q_regr.coef_}}

$$\log\sigma_{t+\Delta} - \log\sigma_{t}=v(W_{t+\Delta}^H - W_{t}^H)$$

## Now we're satisfied we're right, what can we do with it?

````python
def RSFV(H, Delta, F):
    a = (np.cos(H * np.pi) / np.pi) * (Delta ** (H + 0.5))

    t_s = np.arange(len(F), 0, -1)
    b = np.trapz(F / ((t_s + Delta) * (t_s ** (H + 0.5)))) 

    return a * b
````

````python
res = []
T = 1000
for i in range(2, T):
    res.append([i, RSFV(0.14, 10, man_log_var[:i].values)])
````
